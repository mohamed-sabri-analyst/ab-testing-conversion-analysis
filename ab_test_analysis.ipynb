```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

data = pd.DataFrame({
    'user_id': range(1, n+1),
    'group': np.random.choice(['A', 'B'], size=n),
    'purchase': np.random.choice([0,1], size=n, p=[0.85, 0.15]),
    'revenue': np.random.normal(100, 20, size=n)
})

data.loc[data['group']=='B', 'purchase'] = np.random.choice([0,1], size=len(data[data['group']=='B']), p=[0.80,0.20])
```

```python
print(data['revenue'].describe())
```

```python
conversion = data.groupby('group')['purchase'].mean()
print(conversion)
```

```python
from statsmodels.stats.proportion import proportions_ztest

count = data.groupby('group')['purchase'].sum().values
nobs = data.groupby('group')['purchase'].count().values

z_stat, p_value = proportions_ztest(count, nobs)

print(p_value)
```

```python
from scipy.stats import ttest_ind

A_rev = data[data['group']=='A']['revenue']
B_rev = data[data['group']=='B']['revenue']

t_stat, p_value = ttest_ind(A_rev, B_rev)

print(p_value)
```
